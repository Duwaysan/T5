```python
import tensorflow as tf
```

```python
fashion = tf.keras.datasets.fashion_mnist.load_data()
(x_train_val, y_train_val), (X_test, y_test) = fashion
```

```python
x_train_val.shape
```

```python
y_train_val.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.2, random_state=42)
```

```python
X_train_sc, X_val_sc, X_test_sc = X_train / 255.0, X_val / 255.0, X_test / 255.0 #applying scaling (MinMaxScaler)
```

```python
X_train[0] #Using X_train_sc will not show the image because its scaled
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
import numpy as np

X_train_sc.shape
#add another dimention =1
X_train_sc = np.expand_dims(X_train_sc, axis=-1).shape
```

```python
#Building  Arch (How many layers, num units)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Dropout,Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential([Conv2D(20,(3,3),input_shape = (28,28,1)),
                    MaxPooling2D((2,2)),
                    Conv2D(30,(3,3)),
                    MaxPooling2D((2,2)),
                    Conv2D(40,(3,3)),# num of input layer units is defined by the columns automatically
                    MaxPooling2D((2,2)),
                    Flatten(), # Since flatten loses the positions of pixels we'll use it on  Conv2D which in our case build 20 3x3 filter and work on the image
    Dense(units=100, activation='relu'),#kernel_regularizer='l2'),
    Dense(units=10,activation='softmax' )
])

# Compile: Set config
Adam_opt = tf.keras.optimizers.Adam(learning_rate=0.01)
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer='adam', metrics=['acc'],loss='sparse_categorical_crossentropy')

early_Stope = EarlyStopping(monitor='val_loss',patience =8)

#Fit: Search best Paramns(Weight,biases) that minimize loss
history = model.fit(X_train_sc, y_train, epochs=50, validation_data=(X_val_sc, y_val),callbacks=early_stope)


```
