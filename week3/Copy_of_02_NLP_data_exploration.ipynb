#### Most common in Pos vs Neg

```python
pos_tweet = train[train.sentiment == 1]['tweet']
neg_tweet = train[train.sentiment == 0]['tweet']
```

```python
pos_tweet_str = pos_tweet.str.cat(sep = ' ') # make all pos tweets as one string for word cloud
neg_tweet_str = neg_tweet.str.cat(sep = ' ')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

cloud = WordCloud(background_color='white').generate(pos_tweet_str)
plt.imshow(cloud)
plt.axis('off')
plt.title("Positive tweets words");
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

cloud = WordCloud(background_color='black').generate(neg_tweet_str)
plt.imshow(cloud)
plt.axis('off')
plt.title("Negative tweets words");
```
